# Create Subplot

```python
# define figure size
plt.figure(figsize=(10,3))
#create subplot 1 row, 2 col
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))
plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
plt.show()
```

```python
# define figure size
plt.figure(figsize=(10,4))
#create subplot 2 row, 2 col
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.subplot(2,2,4)
plt.plot(x,np.cos(x+1))
plt.show()
```

```python
# define figure size
plt.figure(figsize=(6,5))
#create subplot 2 row, 2 col
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
plt.show()
```

# Legend:
when in plot, label information is there, the same is displayed in plot

```python
plt.figure(figsize=(6,3))
plt.plot(x,np.sin(x),color='blue',label='Sinx')
plt.plot(x,np.cos(x),color='red',label='Cosx')
plt.legend()
plt.show()
```

# 3D Plot
using ax=fig.add_subplot(projection='3d') and ax.scatter()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.random.rand(20)
y=np.random.rand(20)
z=np.random.rand(20)
fig=plt.figure(figsize=(6,3))
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='g',alpha=0.9)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

# Pie Chart (plt.pie)
data, labels, autopct, explode,shadow

```python
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%',explode=(0.05,0.05,0.05),shadow=True)
plt.title("Pie chart")
plt.show()
```

## In Matplotlib:
1. plt.plot() --> Line chart
2. plt.hist() --> Histogram
3. plt.scatter() --> Scatterplot
4. plt.bar() & plt.barh() --> Barplot
5. plt.pie() --> Pie chart
6. ax=fig.add_subplot(projection='3d') and ax.scatter() --> 3d scatter plot

## Univariant Analysis

```python
# For all numerical fields, plot histogram to get idea of frequency
import matplotlib.pyplot as plt
for feature in numerical_cols:
    plt.figure(figsize=(6,3))
    plt.hist(df[feature],bins=20,histtype='barstacked',color='steelblue')
    plt.title(feature + " Plot")
    #plt.axis('equal')
    #plt.legend()
    plt.show()
```

```python
for features in categorical_cols:
    print(f"{features} has {len(df[features].unique())} unique features")
```

```python
# For all categorical fields, plot barplot to get idea of frequency
import matplotlib.pyplot as plt
for feature in categorical_cols:
    if feature=='Surname':
        pass
    else:
        plt.figure(figsize=(6,3))
        plt.bar(df[feature],color='g',height=5)
        plt.title(feature + " Plot")
        #plt.axis('equal')
        #plt.legend()
        plt.show()
```

```python
for feature in categorical_cols:
    if feature=='Surname':
        pass
    else:
        t=[]
        for i in range(len(list(df[feature].unique()))):
            t.append(0.04)
        my_explode=tuple(t)
        
        plt.figure(figsize=(6,3))
        plt.pie(df[feature].value_counts(),explode=my_explode,labels=list(df[feature].unique()),autopct="%1.1f%%",shadow=True)
        plt.title(feature + " Plot")
        #plt.axis('equal')
        #plt.legend()
        plt.show()
```
